```python
#Calculating the VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data=X
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remover(data):
  vif=pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])],index=data.columns)
  if vif.max()>5:
    print(f"Deleting {vif[vif==vif.max()]} for multicollineaity. ")
    data.drop(columns=[vif[vif==vif.max()].index[0]],inplace=True)
  return data
```

```python
for i in range(7):
  vif_data=MC_remover(vif_data)
vif_data.head()
```

```python
vif_data.values
```

```python
vif=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])], index=vif_data.columns)
vif
```

```python
#Train the model
X=vif_data
Y=data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.3,random_state=101)
```

```python
X_train.shape,X_test.shape
```

```python
Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True) # Setting the intercept (i.e., m0) to 0
lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
predictions=lr.predict(X_test)
lr.score(X_test,Y_test)
```

```python
# Verifying the assumptions of the Linear Regression model.
residuals=predictions-Y_test
residualTable=pd.DataFrame({'residuals':residuals,'predictions':predictions})
residualTable=residualTable.sort_values(by='predictions')
```

```python
k=[i for i in range(int(residualTable['predictions'].max()))]
z=[0 for i in range(int(residualTable['predictions'].max()))]
```

```python
plt.figure(figsize=(20,10))
plt.scatter(residualTable['predictions'],residualTable['residuals'],color='red',s=3)
plt.plot(k,z,color='green',linewidth=2,label='regression line')
plt.title('Residual Plot')
plt.xlabel('predictions')
plt.ylim(-800000,800000)
plt.ylabel('residuals w.r.t to predictions')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(residualTable['residuals'],color='red',bins=200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title("Distributions of errors")
plt.show()
```

```python
#Model Coefficients
coefficientTable=pd.DataFrame({'coefficients':lr.coef_,'columns':X_train.columns})
coefficientTable=coefficientTable.sort_values(by='coefficients')
```

```python
plt.figure(figsize=(10,8),dpi=100)
plt.barh(coefficientTable['columns'],coefficientTable['coefficients'])
plt.xlabel('Coefficients')
plt.ylabel('Variables')
plt.title('Normalized Coefficient Plot')
```
